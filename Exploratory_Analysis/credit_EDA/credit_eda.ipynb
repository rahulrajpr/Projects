# **4.Normalising the Data type of the columns**

1. Coversion of Float to Integer

- float columns with null values cannot be converted to integer, because the type of numerical null value is float (Int-null is an invaild object in python)
- float columns without null values can be effectively converted to integer

```python
float_cols_with_nulls = application.select_dtypes(include = np.float).isnull().sum() > 0
float_cols_zero_nulls = application.select_dtypes(include = np.float).isnull().sum() == 0
print('\nCount of float columns with nulls : ',float_cols_with_nulls.sum(),'\n')
print('Count of float columns with zero nulls : ',float_cols_zero_nulls.sum())
```

```python
# Converting float columns with no nulls to Interger

float_cols_zero_nulls_list = application.select_dtypes(include = np.float).columns[float_cols_zero_nulls]

for i in float_cols_zero_nulls_list:
    application[float_cols_zero_nulls_list] = application[float_cols_zero_nulls_list].astype(np.int)

application[float_cols_zero_nulls_list].info() # checking that the float to integer conversion is taken place or not
```

# **6.Binning the Variables ( Numerical to Categorical )**

Variable for Binniing 

* Income to Income levels
* Age to Age Groups

1.Income into the following levels bases on quantiles

* VeryLow - [ 0.0 - 0.1 ]
* Low - [ 0.1 - 0.3 ]
* Mediume - [ 0.3 - 0.6 ]
* High - [ 0.6 - 0.8 ]
* VeryHigh - [ 0.8 - 1.0 ]

```python
# Binning the income uisng the pd.qcut function
application['INCOME_LEVELS'] = pd.qcut(application['AMT_INCOME_TOTAL'], q = [0.0,0.1,0.3,0.6,0.8,1.0], labels = ['VeryLow','Low','Mediume','High','VeryHigh'])
application['INCOME_LEVELS'].value_counts().plot.barh()
plt.show()
```

# **11.Univariate Analysis of Categorical Variables**

Steps involved

1. Creating User Defined Functions
2. Printing summary and visualisation
3. Highlighting inferences

```python
# User Defined function to show the variable summary in tabular form

def col_summary(col_name):
    print('\n','SUMMARY OF',col_name,'\n')
    
    a = application_0[col_name].value_counts().rename('Count T0') # count of non-defaulters
    b = round(application_0[col_name].value_counts(normalize = True)*100).rename('% T0') # percentage of non-defaulters
    
    c = application_1[col_name].value_counts().rename('Count T1') # count of defaulters
    d = round(application_1[col_name].value_counts(normalize = True)*100).rename('% T1') # percentage of defaulters
    
    return pd.concat([a,b,c,d], axis = 1) # printing as dataframe (tabular form)
```

```python
# User Defined function to plot the variable with count plot 

def plot_func(x):

    col_name = x
    title = 'COUNT OF ' + col_name

    fig, axes = plt.subplots(1,2,figsize = [18,6])

    sns.countplot(ax = axes[0], x = application_0[col_name])
    axes[0].set_title('\n'+title+' [ NON DEFAULTER ]'+'\n', color = 'Green', fontsize = 12)
    axes[0].set_xlabel('\n'+col_name+'\n', color = 'red',fontsize = 12)
    axes[0].set_ylabel('COUNT\n', color = 'red',fontsize = 12)
    axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=40, fontsize = 12)

    sns.countplot(ax = axes[1], x = application_1[col_name])
    axes[1].set_title('\n'+title+' [ DEFAULTER ]'+'\n', color = 'Green', fontsize = 12)
    axes[1].set_xlabel('\n'+col_name+'\n', color = 'red',fontsize = 12)
    axes[1].set_ylabel('COUNT\n', color = 'red',fontsize = 12)
    axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=40, fontsize = 12)
    
    plt.show()
```

# **13.Univariate Analysis of Continues Variables**

```python
# User Defined function to show the variable stats in tabular form

def stat_func(col_name):

    print('\nSTATS OF ',col_name,'\n')

    a = application_0[col_name].describe().rename('Non_Defaulter')
    b = application_1[col_name].describe().rename('Defaulter')
    summary = pd.concat([a,b], axis = 1).round(2)
    summary.index.name = 'Target_Type'
    summary = summary.transpose()
    summary['count'] = summary['count'].astype(np.int)
    return summary
```

```python
# User Defined function to plot the variable with count plot 

def plot_func_continues(col_name):

    title = 'DISTRIBUTION OF ' + col_name

    fig, axes = plt.subplots(1,2,figsize = [18,6])
    sns.distplot(ax = axes[0],x = application_0[col_name])
    axes[0].set_title('\n'+title+' [ NON DEFAULTER ]'+'\n', color = 'Green', fontsize = 12)
    axes[0].set_xlabel('\n'+col_name+'\n', color = 'red',fontsize = 12)
    axes[0].set_ylabel('DENSITY\n', color = 'red',fontsize = 12)

    sns.distplot(ax = axes[1],x = application_1[col_name])
    axes[1].set_title('\n'+title+' [ DEFAULTER ]'+'\n', color = 'Green', fontsize = 12)
    axes[1].set_xlabel('\n'+col_name+'\n', color = 'red',fontsize = 12)
    axes[1].set_ylabel('DENSITY\n', color = 'red',fontsize = 12)

    plt.show()
```

# **14.Correlation Analysis Between the Variables**

* On Non-Deafaulters DataFrame

```python
# Linear Correlation Analysis in DataFrame of Non-Defalters [ Target = 0 ]

corr = application_0.corr() # corretaion table

corr_melt = corr.melt(ignore_index = False) # melting the correlation values into single column with keeping in the index inplace
corr_melt = corr_melt.reset_index()

corr_melt.columns = ['VAR1', 'VAR2', 'CORR'] # renamin the columns appropriately

corr_melt['VAR_SET'] = corr_melt[['VAR1','VAR2']].apply(sorted, axis =1) # column containing a sorted list of VAR1 & VAR1, which will help removing duplicates across VAR1 & VAR2 columns

filter_series = corr_melt['VAR_SET'].duplicated(keep = 'first') # boolean series for duplicates with first occurance exception

corr_melt = corr_melt[~filter_series] # removing the duplicates
corr_melt = corr_melt[~(corr_melt['VAR1'] == corr_melt['VAR2'])] # removing same-to-same-value pairs where the correlation value is 1.0, which is an invalid logic
corr_melt = corr_melt[~corr_melt['CORR'].isna()] #removing null rows incase of any

corr_melt['TYPE_CORR'] = corr_melt['CORR'].apply(lambda x : 'Positive' if x > 0 else ('Negative' if x < 0 else 'No Correlation')) # column indicating the direction of correlation
corr_melt['ABS_CORR'] = corr_melt['CORR'].abs() # removing the sign of tge corrlation, so that can be sorted based on the magnitude of the correlation value
top10_corr = corr_melt.sort_values(by = 'ABS_CORR',ascending = False).head(10) # sorting the selecting top10 correction values in descending order in terms of their magintude
top10_corr
```

# **20.Analysis on the Merged DataFrame**

```python
# User Defined function to display the tabular Summary 

def tabular_summary(var1,var2):
    
    p_table = application_merged.pivot_table(index = var1, columns = var2, values = 'SK_ID_CURR', aggfunc = np.size) # creating a pivot_table
    row_sum = p_table.sum(axis = 1) # sum of a row

    for i in p_table.columns:
        p_table[i] = p_table[i]/row_sum # converting to percentage along axis 1

    return p_table
```

```python
# User defined function for stacked bar plot

def stacked_bar_func(var1,var2):

    pv = application_merged.pivot_table(index = var1, columns = var2, values = 'SK_ID_CURR', aggfunc = np.size)

    row_sum = pv.sum(axis = 1)

    for i in pv.columns:
        pv[i] = pv[i]/row_sum

    pv.plot(kind = 'bar', stacked = True, figsize = [12,6])
    plt.title('\n'+var1+' vs '+var2+'%\n', color = 'red', fontsize = 13)
    plt.xlabel('\n'+var1+'\n',color = 'blue', fontsize = 12)
    plt.ylabel('\n'+var2+'%\n',color = 'blue', fontsize = 12)
    plt.show()
```

```python
tabular_summary('FLAG_OWN_CAR','NAME_CONTRACT_STATUS')
```

```python
stacked_bar_func('FLAG_OWN_CAR','NAME_CONTRACT_STATUS')
```

Inference

* Car ownership doesn't have any effect on application approval or rejection. 
* But we saw earlier that the people who has a car has lesser chances of default. 
* The bank can add more weightage to car ownership while approving a loan amount

```python
tabular_summary('CODE_GENDER','NAME_CONTRACT_STATUS')
```

Inference

* Code gender doesn't have any effect on application approval or rejection.
